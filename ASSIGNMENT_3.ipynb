**3. Differences Between np.array(), np.asarray(), and np.asanyarray()**

-   np.array(): Always creates a new array.

-   np.asarray(): Converts input to an array but does not copy if it's
    already an array.

-   np.asanyarray(): Similar to asarray() but allows subclass objects.

**4. Difference Between Deep Copy and Shallow Copy**

-   **Shallow Copy**: Creates a new reference but does not copy nested
    objects.

-   **Deep Copy**: Recursively copies all objects.

Example:

python

CopyEdit

import copy

original = \[\[1, 2, 3\], \[4, 5, 6\]\]

shallow = copy.copy(original)

deep = copy.deepcopy(original)

x = np.random.randn(100)

y = np.random.randn(100)

quadrants = np.where((x \> 0) & (y \> 0), "Q1",

np.where((x \< 0) & (y \> 0), "Q2",

np.where((x \< 0) & (y \< 0), "Q3", "Q4")))
