**1. What is the difference between static and dynamic variables in
Python?**

-   **Static Variables**: In Python, static variables are variables that
    are bound to the class and shared among all instances of that class.
    They retain their value across different instances.

-   **Dynamic Variables**: These variables are bound to an instance and
    can be modified independently for each instance. The value can
    change during runtime.

**Example**:

python

CopyEdit

class MyClass:

static_var = 0 \# Static variable

def \_\_init\_\_(self, value):

self.dynamic_var = value \# Dynamic variable

obj1 = MyClass(10)

obj2 = MyClass(20)

obj1.static_var += 1

print(obj1.static_var, obj2.static_var) \# Static variable is shared

print(obj1.dynamic_var, obj2.dynamic_var) \# Dynamic variable is
instance-specific

**10. What is the difference between deep copy and shallow copy in
Python?**

-   **Shallow Copy**: Creates a new object but does not create copies of
    the nested objects. Instead, it just references the original
    objects.

-   **Deep Copy**: Creates a new object and also recursively copies all
    objects found in the original object, including nested objects.

**Example**:

python

CopyEdit

import copy

original = \[\[1, 2\], \[3, 4\]\]

shallow = copy.copy(original)

deep = copy.deepcopy(original)

original\[0\]\[0\] = 99

print(original) \# Output: \[\[99, 2\], \[3, 4\]\]

print(shallow) \# Output: \[\[99, 2\], \[3, 4\]\] (shallow copy is
affected)

print(deep) \# Output: \[\[1, 2\], \[3, 4\]\] (deep copy is not
affected)

**4. Difference Between Supervised and Unsupervised Learning**

-   **Supervised Learning**:

    -   The model is trained on labeled data (input-output pairs).

    -   The goal is to learn a mapping from inputs to outputs (e.g.,
        classification, regression).

-   **Unsupervised Learning**:

    -   The model is trained on unlabeled data.

    -   The goal is to find hidden patterns or structures in the data
        (e.g., clustering, dimensionality reduction).

**5. Bias-Variance Tradeoff**

-   **Bias**: Error introduced by approximating a real-world problem
    with a simplified model.

-   **Variance**: Error introduced by the model’s sensitivity to small
    fluctuations in the training data.

-   **Tradeoff**: A model with high bias has low variance
    (underfitting), while a model with low bias has high variance
    (overfitting). The goal is to balance both to minimize total error.

**6. Precision and Recall vs Accuracy**

-   **Accuracy**: The proportion of correct predictions (both true
    positives and true negatives) out of all predictions.

-   **Precision**: The proportion of true positives out of all predicted
    positives. It measures how many of the predicted positive cases were
    actually positive.

-   **Recall**: The proportion of true positives out of all actual
    positives. It measures how many of the actual positive cases were
    correctly identified.

**7. Overfitting and How to Prevent It**

-   **Overfitting**: Occurs when a model learns the noise in the
    training data instead of the underlying pattern. It performs well on
    training data but poorly on unseen data.

-   **Prevention**:

    -   Use more training data.

    -   Use regularization (L1, L2).

    -   Prune decision trees.

    -   Use cross-validation.

    -   Simplify the model (e.g., reduce complexity).

**8. Concept of Cross-Validation**

-   Cross-validation is a technique for assessing the generalization
    ability of a model. It involves splitting the data into multiple
    subsets (folds) and training/testing the model on different folds to
    evaluate its performance.

-   Common methods:

    -   **K-Fold Cross-Validation**: The data is split into k folds, and
        the model is trained k times, each time using a different fold
        for testing.

**9. Difference Between Classification and Regression Problems**

-   **Classification**: Predicting categorical labels (e.g., spam or not
    spam).

-   **Regression**: Predicting continuous values (e.g., predicting house
    prices).

**10. Concept of Ensemble Learning**

-   Ensemble learning involves combining multiple models to improve
    performance. The idea is that a group of weak models can make a
    strong model.

-   Common techniques:

    -   **Bagging** (e.g., Random Forest).

    -   **Boosting** (e.g., AdaBoost, XGBoost).

    -   **Stacking**.

**11. Gradient Descent and How It Works**

-   Gradient descent is an optimization algorithm used to minimize the
    cost function of a model by iteratively updating model parameters in
    the direction of the negative gradient.

**12. Batch Gradient Descent vs Stochastic Gradient Descent**

-   **Batch Gradient Descent**: Updates model parameters using the
    entire dataset.

-   **Stochastic Gradient Descent (SGD)**: Updates model parameters
    using a single data point at a time.

**13. Curse of Dimensionality in Machine Learning**

-   The curse of dimensionality refers to the exponential increase in
    computational complexity and data sparsity as the number of features
    (dimensions) increases.

**14. Difference Between L1 and L2 Regularization**

-   **L1 Regularization** (Lasso): Adds the absolute value of the
    coefficients to the cost function. It can result in sparse models
    (some coefficients become zero).

-   **L2 Regularization** (Ridge): Adds the squared value of the
    coefficients to the cost function. It helps prevent overfitting by
    penalizing large coefficients.

**15. Confusion Matrix and Its Use**

-   A confusion matrix is a table used to evaluate the performance of
    classification models. It shows the counts of true positives, true
    negatives, false positives, and false negatives.

**16. AUC-ROC Curve**

-   **AUC** (Area Under the Curve) measures the overall performance of a
    classification model.

-   **ROC Curve** (Receiver Operating Characteristic) plots the true
    positive rate against the false positive rate at various thresholds.

**17. k-Nearest Neighbors (KNN) Algorithm**

-   KNN is a simple, instance-based learning algorithm. It classifies a
    data point based on the majority class of its k nearest neighbors in
    the feature space.

**18. Basic Concept of Support Vector Machine (SVM)**

-   SVM is a supervised learning algorithm used for classification and
    regression. It finds the hyperplane that best separates the data
    into classes.

**19. Kernel Trick in SVM**

-   The kernel trick allows SVM to perform non-linear classification by
    transforming the data into a higher-dimensional space using a kernel
    function.

**20. Types of Kernels in SVM**

-   **Linear Kernel**: Used when data is linearly separable.

-   **Polynomial Kernel**: Used for non-linear data with polynomial
    relationships.

-   **Radial Basis Function (RBF) Kernel**: Commonly used for non-linear
    data.

**21. Hyperplane in SVM and How It Is Determined**

-   A hyperplane is a decision boundary that separates the data into
    different classes. SVM finds the hyperplane that maximizes the
    margin between classes.

**22. Pros and Cons of Using SVM**

-   **Pros**:

    -   Effective in high-dimensional spaces.

    -   Works well for both linear and non-linear data.

-   **Cons**:

    -   Computationally expensive for large datasets.

    -   Sensitive to the choice of kernel and hyperparameters.

**23. Hard Margin vs Soft Margin SVM**

-   **Hard Margin SVM**: Assumes that the data is linearly separable and
    tries to find the optimal separating hyperplane.

-   **Soft Margin SVM**: Allows some misclassification to improve
    generalization when the data is not perfectly separable.

**24. Decision Tree Construction Process**

-   A decision tree is built by recursively splitting the data based on
    the feature that provides the best split (using criteria like Gini
    impurity or information gain).

**25. Working Principle of a Decision Tree**

-   A decision tree makes decisions by following a series of binary
    splits based on feature values. Each node represents a feature, and
    each branch represents a decision rule.

**26. Information Gain and Its Use in Decision Trees**

-   Information gain measures the reduction in entropy (uncertainty)
    achieved by splitting the data based on a feature. It is used to
    select the best feature for a split.

**27. Gini Impurity and Its Role in Decision Trees**

-   Gini impurity is a measure of how often a randomly chosen element
    would be incorrectly classified. It is used to evaluate splits in
    decision trees.

**28. Advantages and Disadvantages of Decision Trees**

-   **Advantages**:

    -   Easy to interpret.

    -   Can handle both numerical and categorical data.

-   **Disadvantages**:

    -   Prone to overfitting.

    -   Sensitive to small changes in the data.

**29. Random Forests and How They Improve Decision Trees**

-   Random forests combine multiple decision trees to improve
    performance. By averaging the predictions of many trees, random
    forests reduce overfitting and variance.

**30. How Random Forest Algorithm Works**

-   Random Forests build multiple decision trees using bootstrapped
    samples of the data. Each tree is trained on a random subset of
    features.

**31. Bootstrapping in Random Forests**

-   Bootstrapping is a sampling technique where random samples are drawn
    with replacement from the original dataset to train each decision
    tree.

**32. Feature Importance in Random Forests**

-   Feature importance measures how useful each feature is in predicting
    the target variable. In Random Forests, it is calculated by looking
    at how much each feature reduces impurity.

**33. Key Hyperparameters of Random Forest**

-   Number of trees (n_estimators).

-   Maximum depth of each tree (max_depth).

-   Minimum samples required to split a node (min_samples_split).

**34. Logistic Regression Model and Its Assumptions**

-   Logistic regression is used for binary classification. It assumes:

    -   The relationship between the independent variables and the
        log-odds of the dependent variable is linear.

    -   The observations are independent.

**35. Logistic Regression for Binary Classification**

-   Logistic regression predicts the probability of a binary outcome
    using the logistic function (sigmoid).

**36. Sigmoid Function in Logistic Regression**

-   The sigmoid function squashes the output of the linear model into a
    probability between 0 and 1.

**37. Cost Function in Logistic Regression**

-   The cost function (log-loss) measures the error between predicted
    probabilities and actual labels. It is minimized during training.

**38. Logistic Regression for Multiclass Classification**

-   Logistic regression can be extended to multiclass classification
    using techniques like **One-vs-All** or **Softmax regression**.

**39. L1 vs L2 Regularization in Logistic Regression**

-   **L1 Regularization**: Can result in sparse models by setting some
    coefficients to zero.

-   **L2 Regularization**: Penalizes large coefficients to prevent
    overfitting.

**40. XGBoost and Its Differences from Other Boosting Algorithms**

-   **XGBoost** is an optimized gradient boosting algorithm that
    improves performance by using regularization, parallelization, and
    tree pruning.

**41. Boosting in Ensemble Learning**

-   Boosting is a technique where weak learners are combined
    sequentially to improve performance. Each new model corrects the
    errors of the previous one.

**42. Handling Missing Values in XGBoost**

-   XGBoost handles missing values by learning the best direction to
    split the data even when some values are missing.

**43. Key Hyperparameters in XGBoost**

-   Learning rate (eta), number of trees (n_estimators), maximum depth
    (max_depth), and subsample ratio (subsample).

**44. Gradient Boosting in XGBoost**

-   Gradient boosting builds trees sequentially, each correcting the
    errors of the previous tree. It uses gradient descent to minimize
    the loss function.

**45. Advantages and Disadvantages of XGBoost**

-   **Advantages**:

    -   High performance and speed.

    -   Handles missing data and outliers well.

-   **Disadvantages**:

    -   Can be prone to overfitting if not tuned properly.

    -   Requires careful hyperparameter tuning.
