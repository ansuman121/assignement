**Web API & Flask Assignment**

**1. What is a Web API?**

A Web API is an interface that allows different software systems to
communicate with each other over the web using HTTP protocols. It
exposes endpoints that enable clients to interact with the server and
perform operations on resources.

**2. How does a Web API differ from a Web Service?**

-   **Web API** is a broader term and can refer to any API that
    communicates over the web using HTTP. It can support multiple data
    formats such as JSON, XML, or HTML.

-   **Web Service** typically refers to SOAP-based services and uses XML
    for communication.

**3. Benefits of using Web APIs:**

-   Enables communication between different applications.

-   Simplifies integration with third-party services.

-   Promotes modularity and scalability in software development.

**4. SOAP vs RESTful APIs:**

-   **SOAP**: A protocol that uses XML for messaging and has strict
    standards for security, transaction compliance, and more.

-   **REST**: An architectural style that uses HTTP methods (GET, POST,
    PUT, DELETE) and supports multiple data formats like JSON and XML.

**5. What is JSON and how is it used in Web APIs?**

JSON (JavaScript Object Notation) is a lightweight data format used for
data exchange between clients and servers. It is easy to read and write
for humans and machines. It is commonly used in RESTful APIs to
represent resources.

**6. HTTP Methods in Web API Development:**

-   **GET**: Retrieve data.

-   **POST**: Create data.

-   **PUT**: Update data.

-   **DELETE**: Remove data.

**7. Authentication & Authorization in Web APIs:**

-   **Authentication**: Verifies the identity of the user (e.g., using
    API keys or OAuth tokens).

-   **Authorization**: Determines what actions the authenticated user
    can perform (e.g., read or write access).

**8. Versioning in Web API Development:**

Versioning ensures backward compatibility when updating an API. It can
be done through URL versioning (e.g., /api/v1/), query parameters, or
headers.

**9. Rate Limiting:**

Rate limiting restricts the number of requests a client can make in a
specified time period to prevent abuse and ensure fair usage.

**10. Statelessness in RESTful APIs:**

Each API request is independent, and the server does not store any
session information. All the necessary information should be provided
with each request.

**11. Error Handling in Web API Responses:**

Error handling is done using HTTP status codes (e.g., 404 for not found,
500 for server error) and providing a detailed error message in the
response body.

**12. Best Practices for Designing and Documenting Web APIs:**

-   Use consistent naming conventions.

-   Provide clear and comprehensive documentation.

-   Version your API to ensure backward compatibility.

-   Use proper HTTP status codes and meaningful error messages.

**13. API Keys and Tokens in Securing Web APIs:**

API keys and tokens are used for authenticating and authorizing API
requests. They help to track usage and ensure that only authorized users
can access the API.

**14. REST Principles:**

-   **Statelessness**: Each request is independent.

-   **Client-Server Architecture**: The client and server are separate
    entities.

-   **Uniform Interface**: A consistent way of interacting with the API.

-   **Resource-based**: Resources are identified by URIs and manipulated
    using HTTP methods.

**15. SOAP vs REST Communication:**

-   **SOAP** uses XML messages with a strict protocol.

-   **REST** uses a flexible, lightweight approach with HTTP methods and
    supports multiple formats like JSON and XML.

**16. Flask Overview:**

Flask is a micro web framework for Python. It is lightweight, simple to
use, and allows for easy routing and template rendering.

**17. Flask Routing:**

Flask uses decorators to map URLs to functions, allowing you to define
routes and their associated handlers.

**18. Flask Templates:**

Flask uses Jinja2 templates to dynamically generate HTML content by
embedding Python code in HTML files.
