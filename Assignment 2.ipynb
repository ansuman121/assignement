**Q1: What is the purpose of the len() function in Python, and how can
it be used with strings?**

**Answer:**

The len() function in Python is used to determine the **length of a
string**, i.e., the number of characters it contains (including spaces
and special characters).

**Example:**

python

CopyEdit

text = "Hello, World!"

print(len(text)) \# Output: 13

Here, the function counts all characters, including spaces and
punctuation, and returns 13.

**Q2: How can you concatenate two or more strings in Python? Can you
provide an example of string concatenation?**

**Answer:**

String **concatenation** in Python can be done using:

1.  The + operator

2.  The .join() method

3.  Using **f-strings** (formatted string literals)

**Examples:**

python

CopyEdit

\# Using + operator

str1 = "Hello"

str2 = "World"

result = str1 + " " + str2

print(result) \# Output: Hello World

\# Using join()

words = \["Python", "is", "awesome"\]

sentence = " ".join(words)

print(sentence) \# Output: Python is awesome

\# Using f-strings

name = "Alice"

greeting = f"Hello, {name}!"

print(greeting) \# Output: Hello, Alice!

**Q3: What is string slicing in Python, and how can you extract a
portion of a string using this method?**

**Answer:**

**String slicing** allows extracting a **substring** from a string using
the format:

python

CopyEdit

string\[start:end:step\]

-   start: The index where slicing begins (inclusive).

-   end: The index where slicing stops (exclusive).

-   step: Optional; defines the increment between indices.

**Examples:**

python

CopyEdit

text = "Python Programming"

\# Extract "Python"

print(text\[0:6\]) \# Output: Python

\# Extract "Programming"

print(text\[7:\]) \# Output: Programming

\# Extract every second character

print(text\[::2\]) \# Output: Pto rgamn

\# Reverse the string

print(text\[::-1\]) \# Output: gnimmargorP nohtyP

**Q4: How can you change the case of a string in Python? Provide
examples of making a string all uppercase and all lowercase.**

**Answer:**

Python provides built-in methods to change the case of a string:

1.  .upper() → Converts all characters to **uppercase**.

2.  .lower() → Converts all characters to **lowercase**.

**Examples:**

python

CopyEdit

text = "Python is Fun!"

\# Convert to uppercase

print(text.upper()) \# Output: PYTHON IS FUN!

\# Convert to lowercase

print(text.lower()) \# Output: python is fun!

**Q5: What does the split() method do in Python, and how can you split a
string using a specific delimiter?**

**Answer:**

The split() method **divides a string into a list of substrings** based
on a specified delimiter. By default, it splits on **whitespace**.

**Examples:**

python

CopyEdit

sentence = "Python is a powerful language"

\# Default split (splitting on spaces)

words = sentence.split()

print(words) \# Output: \['Python', 'is', 'a', 'powerful', 'language'\]

\# Splitting using a specific delimiter

csv_data = "apple,banana,grape,orange"

fruits = csv_data.split(",")

print(fruits) \# Output: \['apple', 'banana', 'grape', 'orange'\]
